```python
import  numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time
```

```python
# id,pw,keyword는 input으로 값을 받는 것으로 수정
url = "https://www.itdaa.net/"

itdaa_id = input()
itdaa_pw = input()

keyword = "비전공자 개발"
```

```python
# 검색 함수
def search_keyword(keyword):
    # 검색어 입력
    driver.find_element_by_id("q").send_keys(keyword)
    # 돋보기 버튼(검색실행) 클릭
    driver.find_element_by_id("Capa_1").click()
    return
```

```python
# 콘텐츠 더보기 클릭 함수
def click_add_btn():
    # 상단 콘텐츠만 더보기 
    while True:
        try:
            add_btn_1 = driver.find_element_by_id("load_more_open_mentorings")
            add_btn_1.click()
            time.sleep(1)
        except:
            break
    # 하단 멘토 콘텐츠 더보기        
    while True:
        try:
            add_btn_2 = driver.find_element_by_id("load_more_mentor_open_mentorings")
            add_btn_2.click()
            time.sleep(1)
        except:
            break
    return
```

```python
# 각 컨텐츠 하이퍼링크 수집
def get_link():
    link_list = []
    try:
        content_link = driver.find_elements_by_css_selector("a.content-body")
        for link in content_link:
            href = link.get_attribute("href")
            link_list.append(href)
        return link_list
    except:
        print("get link error")
```

```python
# 날짜 수집
def get_date():
    date_list = []
    try:
        date = driver.find_elements_by_css_selector("div.created-at")
        for i in date:
            date_list.append(date)
        return date_list

    except:
        print("get date error")
            
```

```python
def get_request_and_parser(url):
    
    request = requests.get(url)
    html_src = request.text
    soup = BeautifulSoup(html_src,'html.parser')    
    
    return soup
    
```

```python
# title,멘티질문,멘토답변 각각 수집하는 함수 구현

def get_title(soup,tag):
    """
    
    tag : 제목 부분에 해당하는 html tag.str타입.2022-03-04 기준으로 "h1"을 입력
    """
    title = soup.find(tag).getText()
#     print(title)
    return title

```

```python
def get_question(soup,tag):
    """

    tag : 멘티질문 부분에 해당하는 html tag.str타입.
    """
    question_text = soup.find("div",{"class":tag}).getText()
    return question_text
```

```python
def get_answer(soup,tag):
    """
    "answer-block"
    """
    answer_text = soup.find("div",{"class":tag}).getText()
    return answer_text
```

```python
url = "https://www.itdaa.net/"
driver = webdriver.Chrome()
driver.get(url)
#로그인버튼 click() 메소드 쓰면 실행되지 않음 왜냐면 button 이 아니고  하이퍼링크가 걸려있음.
# # 아이디와 비밀번호 입력 - 멘토 답변까지 조회하기 위해 로그인이 필요함
try:
    login_link = driver.find_element_by_xpath("""//*[@id="app"]/div[3]/header/div[1]/div/div[2]/div[2]/div/a""") # css selector가 너무 길어 xpath로 접근 
    href = login_link.get_attribute("href")
    # 로그인 페이지 이동
    driver.get(href)
    # 이메일 주소 (아이디) 입력
    user_email = driver.find_element_by_id("user_email")
    user_email.send_keys(itdaa_id)
    # 비밀번호 입력
    user_pw = driver.find_element_by_id("user_password")
    user_pw.send_keys(itdaa_pw)
    # 로그인 버튼 클릭
    login_btn = driver.find_element_by_xpath("""//*[@id="new_user"]/input[2]""")
    login_btn.click()
    
    search_keyword(keyword)
    click_add_btn()
    text_link_list = get_link()
    
    
    date_list = get_date()
    
    title_tag = "h1"
    title_list = []
    
    question_tag = "trix-content froala-content py-4"
    question_text_list = []
    
    answer_tag = "answer-block"
    answer_text_list = []
    
    for url in text_link_list:
        soup = get_request_and_parser(url)
        title = get_title(soup,title_tag)
        question_text = get_question(soup,question_tag)
        answer_text = get_answer(soup,answer_tag)
        
        title_list.append(title)
        question_text_list.append(question_text)
        answer_text_list.append(answer_text)
        
    
except:
    print("error")
# len(title_list),len(question_text_list),len(answer_text_list)
```

```python
pd.DataFrame({"title":title_list,"date":date_list,"question_text":question_text_list,"answer_text":answer_text_list})
```
